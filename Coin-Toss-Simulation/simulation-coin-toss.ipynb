```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Head as 0, Tail as 1
count = 0
count_head = 0
prob_head = []
# Iterate 1000 times (the more, the better)
index = list(range(0,1000))
for x in index:
    toss_value = randint(0,1)
    if toss_value == 0:
        count_head = count_head + 1
    count = count + 1
    probability = count_head / count
    prob_head.append(probability)
```

```python
#Turn into numpy array
np_prob_head = np.array(prob_head)
np_prob_tail = 1 - np_prob_head
```

```python
plt.plot(index,np_prob_head)
plt.show()

plt.plot(index, np_prob_tail)
plt.show()
```
